```python
from PIL import Image
import face_recognition
import os, sys
import cv2
```

```python
# get images from video
def getImagesFromVideo(videoPath, skipFrame, savedImagePath):
    videoData = cv2.VideoCapture(videoPath)
    success,image = videoData.read()
    count = 1
    imgIndex = 1

    while success:
      # skip 
      if count % skipFrame == 0:   
        name = savedImagePath + str(imgIndex) + '.jpg'
        cv2.imwrite(name, image)     # save frame as JPEG file  
        imgIndex +=1
        
      # print('Read a new frame: ', success)
      success,image = videoData.read()
        
      #increment no of frames
      count += 1
    print("Last Image index", imgIndex)
    
    
#rename image files in folder to integers
def renameFiles(imageFolderPath):
    imgIndex = 1
    
    for filename in os.listdir(imageFolderPath):
        os.rename(os.path.join(imageFolderPath, filename), os.path.join(imageFolderPath, str(imgIndex) + '.jpg'))
        imgIndex = imgIndex + 1
        
        
#compress images 
def compressImageSize(startIndex, lastIndex, compressFactor, imageFolderPath, savedImagePath):
    lastIndex += 1
    
    for index in range(startIndex, lastIndex):
        img = Image.open(imageFolderPath + str(index) +".jpg")
        w = int(img.size[0]/compressFactor)
        h = int(img.size[1]/compressFactor)
        img = img.resize((w,h),Image.ANTIALIAS)
        
    #     To rotate saved images
    #     img = img.rotate(-90, expand=True) 
    
        img.save(savedImagePath + str(index) +".jpg", optimize=True, quality=85)
        
        
# face recognition and get coordinates and crop faces
def faceRecognition(startIndex, lastIndex, imageFolderPath, sSavedImgPath, mSavedImgPath, lSavedImgPath):
    lastIndex += 1
    
    for index in range(startIndex, lastIndex):
        image = face_recognition.load_image_file(imageFolderPath + str(index) +".jpg")
        face_locations = face_recognition.face_locations(image)
        print("I found {} face(s) in this photograph.".format(len(face_locations)))

        for face_location in face_locations:
            top, right, bottom, left = face_location
            w = bottom - top
            h = right - left
            print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}, W: {}, H: {}".format(top, left, bottom, right, w, h))

            if w < 100:
                im = Image.open(imageFolderPath + str(index) +".jpg")

                left = left - 20
                if left < 0:
                    left = 0

                top = top-20    
                if top < 0:
                    top = 0

                box = (left, top, right, bottom)
                crop = im.crop(box)
                crop.save(sSavedImgPath + str(index) +'.jpg')
                
            elif w >= 100 and w < 160:
                im = Image.open(imageFolderPath + str(index) +".jpg")
                left = left - 20
                if left < 0:
                    left = 0

                top = top-20    
                if top < 0:
                    top = 0

                box = (left, top, right, bottom)
                crop = im.crop(box)
                crop.save(mSavedImgPath + str(index) +'.jpg')
            else:
                im = Image.open(imageFolderPath + str(index) +".jpg")
                left = left - 20
                if left < 0:
                    left = 0

                top = top-130    
                if top < 0:
                    top = 0

                box = (left, top, right, bottom)

                crop = im.crop(box)
                crop.save(lSavedImgPath + str(index) +'.jpg')
    print("Done!")
```
